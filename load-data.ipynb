```r
library(tidyverse)
library(tidymodels)

data_url <- "https://archive.ics.uci.edu/ml/machine-learning-databases/undocumented/connectionist-bench/sonar/sonar.all-data"
```

```r
data <- read_delim(data_url, delim = ",",
                  col_names=c('FB_01', 'FB_02', 'FB_03', 'FB_04', 'FB_05', 'FB_06', 'FB_07', 'FB_08', 'FB_09', 'FB_10', 'FB_11', 'FB_12', 'FB_13',
            'FB_14', 'FB_15', 'FB_16', 'FB_17', 'FB_18', 'FB_19', 'FB_20', 'FB_21', 'FB_22', 'FB_23', 'FB_24', 'FB_25', 'FB_26',
            'FB_27', 'FB_28', 'FB_29', 'FB_30', 'FB_31', 'FB_32', 'FB_33', 'FB_34', 'FB_35', 'FB_36', 'FB_37', 'FB_38', 'FB_39',
            'FB_40', 'FB_41', 'FB_42', 'FB_43', 'FB_44', 'FB_45', 'FB_46', 'FB_47', 'FB_48', 'FB_49', 'FB_50', 'FB_51', 'FB_52',
            'FB_53', 'FB_54', 'FB_55', 'FB_56', 'FB_57', 'FB_58', 'FB_59', 'FB_60', 'classification')) # FB -> Frequency Band

data
```

```r
data_split <- initial_split(data, prop = 0.75, strata = classification)
    data_training <- training(data_split) #Creates training set

quantity_tbl <- tibble(R_quantity = filter(data_training, classification == "R") %>% nrow(), M_quantity = filter(data_training, classification == "M") %>% nrow())
quantity_tbl #Shows the amount of rows per class

mean_fb_tbl_r <- data_training %>%
                filter(classification == "R") %>%
                select(FB_01:FB_60) %>%
                map_df(mean) %>%
                bind_cols(classification = "R")
#The mean value for each frequency band of class r

mean_fb_tbl_m <- data_training %>%
                filter(classification == "M") %>%
                select(FB_01:FB_60) %>%
                map_df(mean) %>% #no missing data as otherwise this would have given an error
                bind_cols(classification = "M")
#The mean value for each frequency band of class m

mean_fb_tbl <- bind_rows(mean_fb_tbl_r, mean_fb_tbl_m) 
mean_fb_tbl #Shows the mean value for each frequency band by class
```

```r
options(repr.plot.width = 6, repr.plot.height = 10)
data_tall <- gather(data_training, key = FB, value = value, FB_01:FB_60) #Makes it easier to create the following graph:

data_tall_plot <- ggplot(data_tall, (aes(x = FB, y = value))) +
            geom_bar(position = position_dodge(width = 0.7), stat = 'identity', aes(fill = classification)) +
            coord_flip() +
            labs(x = "Frequency Band", y = "Energy Value", fill = "Mine or Rock" ) +
            ggtitle("Energy Value of Each Frequency \nBand by Material") +
            theme(text = element_text(size = 15)) 

data_tall_plot #shows energy value per each frequency band between classes

data_tall_equal <- data_tall %>%
                filter(FB == c("FB_20", "FB_21", "FB_22", "FB_23", "FB_24", "FB_25", "FB_26", "FB_27"))

data_tall_equal_plot <- ggplot(data_tall_equal, (aes(x = FB, y = value))) +
            geom_bar(position = position_dodge(width = 0.7), stat = 'identity', aes(fill = classification)) +
            coord_flip() +
            labs(x = "Frequency Band", y = "Energy Value", fill = "Mine or Rock" ) +
            ggtitle("Energy Value of Each Frequency \nBand by Material") +
            theme(text = element_text(size = 15))

data_tall_equal_plot #zooms in on a few rows that appeared to be equal on t
```
